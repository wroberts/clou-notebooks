```python
gold = {
        'VPC': load_baldwin('intrans.gold') + load_baldwin('trans.gold'),
        'NC': load_nakov('nc_verbs_mturk_first_only.txt'),
    }
```

```python
fns=find_assoc_filenames()
fns
```

```python
all_items = utils.reduce_sets_or(gold.values())
```

```python
lang_filenames = [fn for (l, fn) in fns if l == 'en']
```

```python
filename = lang_filenames[0]
rankings = get_rankings(filename, all_items)
rankings
```

```python
for filename in lang_filenames:
    print(filename)
    rankings = get_rankings(filename, all_items)
    for standard in gold:
        standard_items = set(gold[standard]) & set(rankings)
        print('ranking lists {} out of {} items from gold standard {}'.format(
            len(standard_items), len(gold[standard]), standard))
        standard_rankings = np.array([rankings[item] for item in standard_items])
        print('MRR: {}'.format(eval_mrr(standard_rankings)))
        print('MRR-corr: {}'.format(eval_mrr(standard_rankings,
                                                   len(gold[standard]))))
        print('MAP: {}'.format(eval_map(standard_rankings)))
        print('MAP-corr: {}'.format(eval_map(standard_rankings,
                                                   len(gold[standard]))))
```
